```python
import numpy as np
```

```python
import csv
rows = []
with open('P2_train.csv') as csvfile:
    csvreader = csv.reader(csvfile, delimiter=',')
    for row in csvreader:
        rows.append(row)
    train_data = np.array(rows, dtype = np.float64)
    print(train_data.shape)
rows = []
with open('P2_test.csv') as csvfile:
    csvreader = csv.reader(csvfile, delimiter=',')
    for row in csvreader:
        rows.append(row)
    test_data = np.array(rows, dtype = np.float64)
    print(test_data.shape)
```

```python
#print(train_data[0][2])

train_labels = train_data[:,2]
train_label = train_labels.reshape(train_labels.shape[0], 1)
test_labels = test_data[:,2]
test_label = test_labels.reshape(test_labels.shape[0], 1)
print(train_label.shape)
#train_label[2][0] = 0.0
train_data = train_data[:, :2]
print(train_data.shape)
test_data = test_data[:, :2]
print(test_data.shape)
```

```python
apriori_c0 = float(np.sum(train_label == 0.0))/train_data.shape[0]
apriori_c1 = 1 - apriori_c0
print(train_data.shape[0])
print(apriori_c0)
print(apriori_c1)
```

```python
n0 = np.sum(train_labels == 0.0)
n1 = train_data.shape[0] - n0
print(n0, n1)
```

```python
bool_index = (train_labels == 0.0)
print(bool_index.shape)
#mu_0 = np.sum(train_data, axis = 0, keepdims = True)
#mu_5.shape
```

```python
train_data0 = np.zeros([n0, train_data.shape[1]])
train_data1 = np.zeros([n1, train_data.shape[1]])
j = 0; k=0
for i in range(train_data.shape[0]):
    if bool_index[i] == True:
        train_data0[j] = train_data[i]
        j +=1
    else: 
        train_data1[k] = train_data[i]
        k +=1
print(train_data0.shape)
#train_data5[394]
#train_data6.shape
```

```python
# DIVIDE TRAINING DATA IN 2 MATRICES : train_data5 and train_data6
# Let the number of examples in each be n5 and n6
mu_0 = np.sum(train_data0, axis = 0, keepdims = True)/n0
#mu_5.shape
mu_1 = np.sum(train_data1, axis = 0, keepdims = True)/n1
#mu_5.shape
```

```python
sigma_0 = np.dot((train_data0 - mu_0).T, (train_data0 - mu_0))/(n0-1)
sigma_1 = np.dot((train_data1 - mu_1).T, (train_data1 - mu_1))/(n1-1)
print(sigma_1)
```

```python
# Defining the discriminant functions g5 and g6
predict_label = []
g0 = np.zeros([test_data.shape[0], 1])
g1 = np.zeros([test_data.shape[0], 1])
for i in range(test_data.shape[0]):
    g0[i] = (-1/2)*np.log10(np.linalg.det(sigma_0))+ (-1/2)*np.dot(np.dot(test_data[i]-mu_0, np.linalg.inv(sigma_0)), (test_data[i]-mu_0).T) + np.log10(apriori_c0)
    g1[i] = (-1/2)*np.log10(np.linalg.det(sigma_1)) + (-1/2)*np.dot(np.dot(test_data[i]-mu_1, np.linalg.inv(sigma_1)), (test_data[i]-mu_1).T)+ np.log10(apriori_c1)
    if g1[i] >= g0[i]:
        predict_label.append(1.0)
    else:
        predict_label.append(0.0)
predictions = np.array(predict_label)
predictions = predictions.reshape(test_data.shape[0], 1)
```

```python
test_c0 = sum(test_label == 0.0)
test_c1 = test_label.shape[0] - test_c0
infer = (predictions == test_label)
accuracy = (float(sum(infer))/test_data.shape[0]) * 100
print('The accuracy of the Bayesian classifier is %f '%(accuracy) + '%')
#print(infer[0:20])
error0 = 0
for i in range(test_label.shape[0]):
    if test_label[i][0] == 0.0 and predictions[i] == 1.1:
        error0 += 1
misclass_c0 = (float(error0)/test_c0) * 100
print('The misclassification error for class 0 is %f '%(misclass_c0) + '%')
error1 = 0
for i in range(test_label.shape[0]):
    if test_label[i][0] == 1.0 and predictions[i] == 0.0:
        error1 += 1
misclass_c1 = (float(error1)/test_c1) * 100
print('The misclassification error for class 1 is %f '%(misclass_c1) + '%')
```
